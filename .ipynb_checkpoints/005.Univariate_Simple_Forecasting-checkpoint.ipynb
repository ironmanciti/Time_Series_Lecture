# Univariate Simple Forecasting

## Grid Search 를 하여 best result 를 얻는 parameter 를 찾음

```python
import numpy as np
from sklearn.metrics import mean_squared_error

def simple_forecast(history, config):   # config (n 개의 timestep, window size, forecast type)
    n, offset, avg_type = config
    
    if avg_type == 'persist':
        return history[-n]
    
    values = []
    
    if offset == 1:
        values = history[-n:]
    else:
        if n * offset > len(history):
            raise Exception("Config beyond end of data: {}".format(n, offset))
            
        for i in range(1, n):
            ix = i * offset
            values.append(history[-ix])

    if avg_type == 'mean':
        return np.mean(values)
    elif avg_type == 'median':
        return np.median(values)
    else:
        raise Exception("Invalid avg_type: {}".format(avg_type))
        
def rmse(actual, predicted):
    return np.sqrt(mean_squared_error(actual, predicted))
        
def train_test_split(data, n_test):      # n_test 개를 test data 로 split
    return data[:-n_test], data[-n_test:]

def walk_forward_validation(data, n_test, cfg):
    predictions = []
    train, test = train_test_split(data, n_test)
    history = [x for x in train]
    
    for i in range(len(test)):
        yhat = simple_forecast(history, cfg)
        predictions.append(yhat)
        history.append(test[i])
        
    error = rmse(test, predictions)
    return error

def score_model(data, n_test, cfg, debug=False):
    result = None
    key = cfg
    
    result = walk_forward_validation(data, n_test, cfg)
    
    if debug and result is not None:
        print(" > Model[{}] {}".format(key, result))
    return (key, result)

def grid_search(data, cfg_list, n_test, pararell=True):
    scores = [score_model(data, n_test, cfg) for cfg in cfg_list]
    scores = [r for r in scores if r[1] != None]
    scores.sort(key=lambda tup: tup[1])
    return scores

def simple_configs(max_length, offsets=[1]):
    configs = []
    for i in range(1, max_length+1):
        for o in offsets:
            for t in ['persist', 'mean', 'median']:
                cfg = [i, o, t]
                configs.append(cfg)
    return configs
```

```python
data = [10.0, 20.0, 30.0, 40.0, 50.0, 60.0, 70.0, 80.0, 90.0, 100.0]
n_test = 4
max_length = len(data) - n_test
cfg_list = simple_configs(max_length)
scores = grid_search(data, cfg_list, n_test)
print('done')
for cfg, error in scores:
    print(cfg, error)
```
