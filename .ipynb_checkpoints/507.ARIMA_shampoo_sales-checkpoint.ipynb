# ARIMA

- stationary 한 dataset 에 적용  

- non-seasonal ARIMA : ARIMA(p, d, q)  $\rightarrow$ 3 개 parameter 학습

- seasonal ARIM : ARIMA(p, d, q)(P, D, Q)m  $\rightarrow$ 7 개 parameter 학습  

- AR (AutoRegressive) $\rightarrow$ p  
- I (Integrated) $\rightarrow$ d  
- MA (Moving Average) $\rightarrow$ q 

## Shampoo Sales Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

squeeze : bool, default False, If the parsed data only contains one column then return a Series.

```python
def parser(x):
    return pd.datetime.strptime('190' + x, '%Y-%m')

series = pd.read_csv('shampoo-sales.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
```

```python
series[:5]
```

```python
series.plot()
```

- shampoo data 는 trend 가 확실히 보이므로 non-stationary 하므로 stationary 하게 만들기 위해 differencing 필요. d = 1

- ARIMA(p, d, q)  
    p – lagged observation number (lag order)  
    d – degree of differencing  
    q – window size  

- parameter p 를 구하기 위해 pandas built-in 기능을 이용하여 autocorrelation 을 plot.

```python
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(series)
```

- lag 10 이전까지 positive correlation 을 보이고 lag 5 이전까지 확실히 보이므로 p = 5 로 설정


- statsmodel library 를 이용하여 ARIMA model 생성  
    (1) ARIMA() 호출  
    (2) fit() 함수로 training  
    (3) predict() 함수로 예측  

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(series, order=(5, 1, 0))
model_fit = model.fit(disp=0)   # convergence information is not printed
print(model_fit.summary())
```

plot residual errors

```python
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
residuals.plot(kind='kde')
print(residuals.describe())
```

### Rolling Forecast ARIMA model

- predict() 함수를 이용한 prediction

```python
X = series.values
print(X.shape)

train_size = int(X.shape[0] * 0.66)
train, test = X[:train_size], X[train_size:]
print(train.shape)
print(test.shape)
```

### forecast() function performs a one-step forecast using the model.

```python
history = [x for x in train]
predictions = []

for t in range(len(test)):
    model = ARIMA(history, order=(5, 1, 0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
    
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
plt.plot(test, label="expected value")
plt.plot(predictions, color='red', label="rolling forecast predictions")
plt.legend()
```
