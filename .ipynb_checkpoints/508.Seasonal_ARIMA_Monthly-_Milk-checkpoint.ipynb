# Seasonal ARIMA - Monthly Milk Production

Seasonal ARIM : ARIMA(p, d, q)(P, D, Q)m → 7 개 parameter 학습

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("monthly-milk-production.csv", index_col="Month", parse_dates=True)
df.head()
```

```python
df.info()
```

```python
df.plot()
```

육안으로 보면 trend 와 seasonality 보임

```python
timeseries = df['Milk Production']

timeseries.rolling(12).mean().plot()
```

mean 은 non-stationary

```python
timeseries.rolling(12).std().plot()
```

variance (std * std) 역시 fluctuation 이 크므로 time invariant $\rightarrow$ non-stationary

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

decomposition = seasonal_decompose(timeseries)

decomposition.plot();
```

seasonality 와 trend 가 확인됨

```python
def adf_check(ts):
    result = adfuller(ts)
    print('ADF Statistics: {:f}'.format(result[0]))
    print('p-value: {:f} - {}'.format(result[1], 'Non-stationary' if result[1] > 0.05 else 'Stationary'))
    print('Critical Values:')
    for k,v in result[4].items():
        print('\t{}: {:.3f}'.format(k, v))
```

```python
adf_check(timeseries)
```

p-value > 0.05 이므로 non-stationary

## Differencing

- 1 차 differencing

```python
df['1st diff'] = df['Milk Production'] - df['Milk Production'].shift(1)
df.head()
```

```python
adf_check(df['1st diff'].dropna())
```

```python
df['1st diff'].plot()
```

2 차 differencing

```python
df['2nd diff'] = df['1st diff'] - df['1st diff'].shift(1)
df.head()
```

```python
adf_check(df['2nd diff'].dropna())
```

```python
df['2nd diff'].plot()
```

## Seasonal Differencing

```python
df['seasonal diff'] = df['Milk Production'] - df['Milk Production'].shift(12)
df['seasonal diff'].plot()
```

```python
adf_check(df['seasonal diff'].dropna())
```

seasonal differencing method 2 $\rightarrow$ 1st differencing 에 seasonal differencing

```python
df['seasonal diff2'] = df['1st diff'] - df['1st diff'].shift(12)
df['seasonal diff2'].plot()
```

```python
adf_check(df['seasonal diff2'].dropna())
```

### 위의 결과에 의해 I(d) parameter 결정 : d = 1, D = 1

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(df['1st diff'].dropna());
```

```python
plot_acf(df['seasonal diff2'].dropna());
```

```python
plot_pacf(df['1st diff'].dropna(), method='ywm');
```

```python
plot_pacf(df['seasonal diff2'].dropna(), method='ywm');
```

### 위 결과에 따라 P = 1, Q = 1, p=0, q=0 로 정하여 training 한 후 parameter 개선

(p, d, q) = (0, 1, 0)  
(P, D, Q)m = (1, 1, 1) 12

```python
model = sm.tsa.statespace.SARIMAX(df['Milk Production'], order=(0, 1, 0), seasonal_order=(1, 1, 1, 12))

result = model.fit()

print(result.summary())
```

training 중 계산된 residual

```python
result.resid.plot();
```

```python
result.resid.plot(kind='kde');
```

```python
len(df)
```

```python
df['forecast'] = result.predict(start=140, end=168)

df[['Milk Production', 'forecast']].plot(figsize=(8, 6))
```
