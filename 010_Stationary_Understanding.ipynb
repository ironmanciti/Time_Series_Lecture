# Stationary (정상성) Time Series

```python
import statsmodels as ss
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## What is Stationarity?

1. constant mean (일정한 평균)  
2. constant variance (일정한 분산)  
3. constant autocorrelation structure (일정한 자기상관)  
4. no periodic component (주기 성분 없음)

```python
np.random.seed(101)

time = np.arange(500)
stationary = np.random.normal(loc=0, scale=1.0, size=len(time))
```

```python
def plot_sequence(x, y, title, xlabel="time", ylabel="series"):
    plt.figure(figsize=(16, 4))
    plt.plot(x, y, 'k-')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(alpha=0.3);
```

```python
plot_sequence(time, stationary, title="Stationary Time Series")
```

- 위의 플롯에서 뚜렷한 추세나 계절성이 없는 것에 주목하십시오. 이 시계열 데이터는 4가지 정상성 조건인 일정한 평균, 일정한 분산, 일정한 자기상관, 주기 성분 없음을 충족하는 것으로 보입니다.  

- 시계열의 histogram 을 plotting 하여 정규분포를 보이면 평균과 분산이 일정하다는 확신을 얻을 수 있습니다.

```python
pd.Series(stationary).hist();
```

## Autocorrelation Structure

- 자기상관성은 difference(차분)에 의해 stationary 로 바꿀 수 있다.

```python
# seed to start series
seed = 3.14

# create autocorrelated data
lagged = np.empty_like(time, dtype='float')
for t in time:
    lagged[t] = seed + np.random.normal(loc=0, scale=2.5, size=1)
    seed = lagged[t]
```

```python
plot_sequence(time, lagged, title="Nonstationary Data w/Lagged Structure")
```

시계열이 균일분포 등 비정규분포를 보이면 non-stationay 라고 판단 가능

```python
pd.Series(lagged).hist();
```

## non-stationary (비정상) data

1. Trend (평균 변화)  
2. 이분산성(변동 분산)  
3. Seasonality (계절성, 주기적 구성요소)  
4. Trend + Seasonality

### Trend

평균은 시간이 지남에 따라 변하므로 non-stationary

```python
trend = (time * 2.75) + stationary

plot_sequence(time, trend, title="Nonstationary Data w/Trend")
```

### 이분산성(변동 분산)

시간에 따른 분산의 변화

```python
np.random.seed(1234)

# data
level_1 = np.random.normal(loc=0, scale=1.0, size = 250)
level_2 = np.random.normal(loc=0, scale=10.0, size = 250)
data = np.append(level_1, level_2)

plot_sequence(time, data, title="Nonstationary w/Varying variance")
```

### Seasonality

주기적인 요소가 있으므로 non-stationary

```python
seasonality = 10 + np.sin(time) * 10

plot_sequence(time, seasonality, title="Nonstationary Data w/Seasonality")
```

### Trend + Seasonality

평균은 시간이 지남에 따라 변하고 주기적인 구성 요소가 있습니다.

```python
trend_seasonality = trend + seasonality + stationary

plot_sequence(time, trend_seasonality, title="Nonstationary Data w/Trend + Seasonality")
```

### Differencing 으로 Autocorrelation 제거

```python
difference = lagged[:-1] - lagged[1:]

plot_sequence(time[:-1], difference, title="Stationary Data After Differencing")
```
