# Persistence algorithm 을 이용한 forecasting 

- Persistence Algorithm : last value 를 next value 로 예측에 사용


- 약간의 noise 를 추가한 seasonality 와 trend 가 있는 time series 이용 

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.metrics import mean_squared_error, mean_absolute_error
np.random.seed(101)
````

````python
def seasonal_pattern(season_time):
    """Just an arbitrary pattern, you can change it if you wish"""
    return np.where(season_time < 0.4,
                    np.cos(season_time * 2 * np.pi),
                    1 / np.exp(3 * season_time))

def seasonality(time, period, amplitude=1, phase=0):
    """Repeats the same pattern at each period"""
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)

time = np.arange(4 * 365 + 1, dtype="float32")   # 4 년
baseline = 10          # 기본값
series = 0.1 * time    # trend 시계열 data 
amplitude = 40
slope = 0.05
noise_level = 5

# series data 생성
series = baseline + 0.05 * time + \
         seasonality(time, period=365, amplitude=amplitude) + \
         np.random.randn(len(time))  # noise 추가

plt.figure(figsize=(8, 5))
plt.plot(series)
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
plt.show()
````

````python
series.shape
````

### 위에서 생성한 time series 를 train 와 validation set 으로 분할

````python
split_time = 1000

time_train = time[:split_time]
x_train = series[:split_time]

time_valid = time[split_time:]
x_valid = series[split_time:]

plt.figure(figsize=(8, 5))
plt.plot(time_train, x_train)
plt.plot(time_valid, x_valid)
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
plt.show()
````

# Naive Forecast

- last value 를 prediction 으로 사용 -> one-step behind prediction

````python
naive_forecast = series[split_time - 1:-1]
naive_forecast[:10]
````

````python
x_valid[:10]
````

````python
plt.figure(figsize=(8, 4))
plt.plot(time_valid, x_valid)
plt.plot(time_valid, naive_forecast)
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
````

naive forecast 가 one-step behind 인지 그래프상으로 구별 어려우므로 validation period 의 앞부분으로 zoom in.

````python
plt.figure(figsize=(10, 6))
plt.plot(time_valid, x_valid)
plt.plot(time_valid, naive_forecast)
plt.xlabel("Time")
plt.ylabel("Value")
plt.xlim(1000, 1101)
plt.grid(True)
````

- naive forecast 가 1 step behind 지연되어 time series 를 그대로 재현하고 있음

### naive forecast 와 validation period 값 간의 차이를  mse 와 mae 로 계산

````python
print(mean_squared_error(x_valid, naive_forecast))
print(mean_absolute_error(x_valid, naive_forecast))
````

# Real Data를 이용한 Naive Forecasting

`datetime.strptime` : 주어진 문자열에서 datetime 객체를 생성합니다.

```
"Month","Sales"
"1-01",266.0
"1-02",145.9
"1-03",183.1
"1-04",119.3
```

````python
def parser(x):
    return datetime.strptime('190' + x, '%Y-%m')
````

````python
series = pd.read_csv("datasets/shampoo-sales.csv", parse_dates=[0], 
                     date_parser=parser, index_col=0, header=0)
series.head()
````

````python
series.plot()
````

### Train / Test set 분리

````python
split_time = int(len(series) * 0.6)

X_train = series[:split_time]

X_valid = series[split_time:]

plt.figure(figsize=(8, 5))
plt.plot(series.index[:split_time], X_train)
plt.plot(series.index[split_time:], X_valid)
plt.xlabel("Time")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
````

### Persistence Algorithm define

````python
def model_persistence(series, split_time):
    return series[split_time-1:-1]
````

### Forecast and Evaluation

````python
prediction = model_persistence(series, split_time)

test_score = mean_squared_error(X_valid, prediction)
print('Test MSE: {:.3f}'.format(test_score))
````

````python
plt.figure(figsize=(8, 5))
plt.plot(series.index[:split_time], X_train, label='train')
plt.plot(series.index[split_time:], X_valid, label='valid')
plt.plot(series.index[split_time:], prediction, label='prediction')
plt.xlabel("Time")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
````
